```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
from numpy import *
from sklearn.linear_model import LinearRegression
```

```python
points = pd.read_csv('data.csv')

print(points.shape)

x=points.iloc[:,0].values

y=points.iloc[:,1].values

#Plot the dataset
plt.scatter(x,y)
plt.xlabel('Hours of study')
plt.ylabel('Test scores')
plt.title('Dataset')
plt.show()


x = np.reshape(x, (99, -1))
y=np.reshape(y, (99, -1))

print(x.shape)
```

```python
reg = LinearRegression()
reg.fit(x,y)



```

```python
# Y Prediction
print ("******************************")
print (reg.coef_)
print (reg.intercept_)
```

```python
print(reg.predict([[50]]))
```

```python
plt.scatter(x,y)
plt.xlabel('Hours of study')
plt.ylabel('Test scores')
plt.title('Dataset')

plt.plot(x, reg.coef_*x + reg.intercept_,"r")

plt.show()
```
